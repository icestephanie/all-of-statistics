```python
import numpy as np
from scipy.stats import bernoulli
from tqdm.notebook import tqdm

np.random.seed(0)
alpha=0.05
p=0.4

B=50000
N=10000

nn=np.arange(1,N+1)
epsilon_n=np.sqrt((1/(2*nn))*np.log(2/alpha))


p_hat=np.cumsum(bernoulli.rvs(p, size=(B,N), random_state=np.random.randint(N)),axis=1)
    
coverage=np.mean((p_hat+epsilon_n>=p) & (p_hat-epsilon_n<=p), axis=0)
p_hat.shape
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(12,8))
plt.plot(nn, coverage)
plt.xlabel('n')
plt.ylabel('coverage')
plt.show()
```
